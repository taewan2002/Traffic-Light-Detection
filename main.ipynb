```python
import os
import shutil
import random

# ÎîîÎ†âÌÜ†Î¶¨ Í≤ΩÎ°ú ÏÑ§Ï†ï
train_images_dir = 'train/images'
train_labels_dir = 'train/labels'
valid_images_dir = 'valid/images'
valid_labels_dir = 'valid/labels'

# Ïú†Ìö®ÏÑ± Ìè¥Îçî ÏÉùÏÑ±
os.makedirs(valid_images_dir, exist_ok=True)
os.makedirs(valid_labels_dir, exist_ok=True)

# train/labels Ìè¥ÎçîÏùò ÌååÏùº Î™©Î°ùÏùÑ Í∞ÄÏ†∏ÏòµÎãàÎã§.
label_files = os.listdir(train_labels_dir)

# ÌååÏùº Î™©Î°ùÏùò 20%Î•º Î¨¥ÏûëÏúÑÎ°ú ÏÑ†ÌÉùÌï©ÎãàÎã§.
num_files_to_move = int(0.2 * len(label_files))
label_files_to_move = random.sample(label_files, num_files_to_move)

# ÏÑ†ÌÉùÎêú ÌååÏùºÎì§Í≥º ÎåÄÏùëÎêòÎäî Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏùÑ valid Ìè¥ÎçîÎ°ú Ïù¥ÎèôÌï©ÎãàÎã§.
for label_file in label_files_to_move:
    label_src_path = os.path.join(train_labels_dir, label_file)
    label_dest_path = os.path.join(valid_labels_dir, label_file)
    
    # Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏùò ÌôïÏû•Ïûê Ï∂îÎ°†
    image_file = label_file.replace('.txt', '.jpg')  # Ïòà: .jpg, .png, .jpeg Îì±
    image_src_path = os.path.join(train_images_dir, image_file)
    image_dest_path = os.path.join(valid_images_dir, image_file)
    
    # ÎùºÎ≤® ÌååÏùº Ïù¥Îèô
    shutil.move(label_src_path, label_dest_path)
    print(f"Moved label: {label_file}")
    
    # ÎåÄÏùëÎêòÎäî Ïù¥ÎØ∏ÏßÄ ÌååÏùº Ïù¥Îèô
    if os.path.exists(image_src_path):
        shutil.move(image_src_path, image_dest_path)
        print(f"Moved image: {image_file}")
    else:
        print(f"Image file not found for label: {label_file}")

print(f"Moved {num_files_to_move} label and image files from {train_labels_dir} to {valid_labels_dir} and {valid_images_dir}")
```
